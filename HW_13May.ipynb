```python
class Exchange:
  def __init__(self, rate):
    self.rate = rate

  def convert(self, amount, from_currency, to_currency):
    amount_in_byn = amount * self.rate[from_currency]
    converted_amount = amount_in_byn / self.rate[to_currency]
    return round(converted_amount, 2)

class CurrencyAccount:
  def __init__(self, balance, currency):
    self._balance = balance
    self.currency = currency

  @property
  def balance(self):
    return self._balance

  def add(self, amount):
    self._balance += amount

  def deduct(self, amount):
    if self._balance < amount:
      print(f'Недостаточно средств')
      return False
    else:
      self._balance -= amount
      return True

class BankAccount:
  def __init__(self, owner, accounts):
    self.owner = owner
    self.accounts = accounts

  def internal_transfer(self, from_account_name, to_account_name, amount):
    if from_account_name in self.accounts and to_account_name in self.accounts:
      from_account = self.accounts[from_account_name]
      to_account = self.accounts[to_account_name]
      if from_account.deduct(amount):
        if from_account.currency == to_account.currency:
          converted_amount = amount
        else:
            exchanger = Exchange(rates)
            converted_amount = exchanger.convert(amount, from_account.currency, to_account.currency)
      to_account.add(converted_amount)
      print(f'Перевод выполнен успешно')
    else:
      print(f'Выбранный счет не найден')

  def external_transfer(self, external_account, my_account_name, external_account_name, amount):
    if my_account_name in self.accounts and external_account_name in external_account.accounts:
      my_account = self.accounts[my_account_name]
      external_account = external_account.accounts[external_account_name]
      if my_account.deduct(amount):
        if my_account.currency == external_account.currency:
          converted_amount = amount
        else:
            exchanger = Exchange(rates)
            converted_amount = exchanger.convert(amount, my_account.currency, external_account.currency)
      external_account.add(converted_amount)
      print(f'Перевод выполнен успешно')
    else:
      print('Выбранный счет не найден')

rates = {"USD": 2.81, "EUR": 3.30, "BYN": 1.0}

my_usd_account = CurrencyAccount(1000, "USD")
my_eur_account = CurrencyAccount(33, "EUR")
my_byn_account = CurrencyAccount(700, "BYN")
my_byn2_account = CurrencyAccount(10, "BYN")

my_accounts_dict = {"USD": my_usd_account,"EUR": my_eur_account,"BYN1": my_byn_account, "BYN2": my_byn2_account}
my_account = BankAccount("Yana", my_accounts_dict)

external_usd_account = CurrencyAccount(666, "USD")
external_usd2_account = CurrencyAccount(20, "USD")
external_eur_account = CurrencyAccount(25, "EUR")
external_byn_account = CurrencyAccount(2000, "BYN")

external_accounts_dict = {"USD1": external_usd_account,"USD2": external_usd2_account, "EUR": external_eur_account,"BYN": external_byn_account}
external_bank_account = BankAccount("Maxim", external_accounts_dict)

my_account.internal_transfer("BYN1", "BYN2", 100)
my_account.internal_transfer("BYN2", "USD", 10)

print(f'Баланс внутреннего счета BYN после внутреннего перевода: {my_byn_account.balance}')
print(f'Баланс внутреннего счета BYN2 после внутреннего перевода: {my_byn2_account.balance}')
print(f'Баланс внутреннего счета USD после внутреннего перевода: {my_usd_account.balance}')

my_account.external_transfer(external_bank_account, "USD", "USD1", 100)
my_account.external_transfer(external_bank_account, "USD", "BYN", 10)

print(f'Баланс внутреннего счета USD после внешнего перевода: {my_usd_account.balance}')
print(f'Баланс внешнего счета USD1 после внешнего перевода: {external_usd_account.balance}')
print(f'Баланс внешнего счета BYN после внешнего перевода: {external_byn_account.balance}')















```
