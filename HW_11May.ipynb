```python
# программа, которая запрашивает у пользователя текст и выводит:
# количество слов,количество уникальных слов,
# самое часто встречающееся слово (и сколько раз),
# среднюю длину слова.

from collections import Counter

my_text = input("Введите желаемый текст: ")
my_text = my_text.lower()

my_words = (len(my_text.split()))
print ("Кол-во слов: ",my_words)

my_wordsun = len(list(dict.fromkeys(my_text.split())))
print ("Кол-во уникальных слов: ",my_wordsun)

my_often = Counter(my_text.split()).most_common(1)
print ("Самое частое слово и его кол-во повторений: ",my_often)

my_split = my_text.split()

lengh = 0
for i in range(len(my_split)):
  lengh += len(my_split[i])
avg_len = lengh / (len(my_text.split()))
print("Средняя длина слова: ", avg_len)
```

```python
# Компьютер загадывает число от 1 до 100. Пользователь пытается угадать. 
# После каждой попытки компьютер говорит «больше», «меньше» или «угадал». В конце вывести количество попыток.

import random

comp_var = random.randint(0, 100)
user_var = int(input('Какое число загадал компьютер от 0 до 100:'))

while user_var != comp_var:
  if user_var > comp_var:
    print('Меньше')
    user_var = int(input('Какое число загадал компьютер:'))
  elif user_var < comp_var:
    print('Больше')
    user_var = int(input('Какое число загадал компьютер:'))
print('Угадал')
```

```python
# Напишите программу «Таймер». Пользователь вводит количество секунд. 
# Программа отсчитывает время, каждую секунду выводя оставшееся время в формате MM:SS.

import time

input_seconds = int(input('Введите количество секунд: '))
start_timer = input_seconds

while start_timer >= 0:
  minutes = start_timer // 60
  seconds = start_timer % 60
  print('{:02d}:{:02d}'.format(minutes, seconds))
  time.sleep(1)
  start_timer -= 1
```

```python
#Write a method that takes an array of consecutive (increasing) letters
#as input and that returns the missing letter in the array

import string
def find_missing_letter(chars):
    first_char = chars[0]
    up_low = first_char.isupper()
    if up_low:
        alphabet = string.ascii_uppercase
    else:
        alphabet = string.ascii_lowercase
    chars = list(chars)
    start_index = alphabet.index(chars[0])
    chars_list = alphabet[start_index : start_index + len(chars) + 1]
    for letter in chars_list:
        if letter not in chars:
            return letter
find_missing_letter(['d', 'e', 'g', 'h'])
```

```python
#Assume "#" is like a backspace in string. This means that string "a#bc#d" actually is "bd"
#Your task is to process a string with "#" symbols

def clean_string(s):
    if s == None:
        return ""
    clean_list = []
    for letter in s:
        if letter != '#':
            clean_list.append(letter)
        else:
            if clean_list:
                clean_list.pop()
    return ''.join(clean_list)
clean_string('abc#d##c')
```

```python
# Complete the solution so that it splits the string into strings of
# two characters in a list.If the string contains an odd number of
# characters then it should replace the missing second character of
# the final pair with an underscore ('_').

def solution(s):
    sep_list = []
    if len(s) % 2 == 0:
        for i in range(0, len(s), 2):
          sep_list.append(s[i:i+2])
    else:
      s = s + '_'
      for i in range(0, len(s), 2):
        sep_list.append(s[i:i+2])
    return sep_list
solution('abcdef')
```

```python
# Move the first letter of each word to the end of it,
# then add "ay" to the end of the word.
# Leave punctuation marks untouched.

import string

def pig_it(text):
    pig_list = []
    pig_text = text.split()
    for pig in pig_text:
        if pig.isalpha():
            pig_list.append(pig[1:] + pig[0] + 'ay')
        else:
            pig_list.append(pig)
    return ' '.join(pig_list)
pig_it('Pig latin is cool')
```

```python
# Create hashtag generator. It must start with a hashtag (#).
# All words must have their first letter capitalized,
# and remaining letters lowercased.
# If the final result is longer than 140 chars it must return false.
# If the input or the result is an empty string it must return false.

def generate_hashtag(s):
    if not s.strip():
        return False
    else:
        s = s.title()
        s = s.replace(' ', '')
    if len(s) + 1 > 140:
        return False
    else:
        return ('#' + s)
generate_hashtag('c i n')
```

```python
# Complete the function that returns true
# if a portion of str1 characters can be rearranged to match str2,
# otherwise returns false.

from collections import Counter

def scramble(s1, s2):
  s1_counts = Counter(s1)
  s2_counts = Counter(s2)

  return (s1_counts & s2_counts) == s2_counts

print(scramble('rkqodlw', 'world'))
```

```python
# the board comes in the form of a 3x3 array, where the value is 0,
# if a spot is empty, 1 if it is an "X", or 2 if it is an "O"
# function will return -1 if the board is not yet finished AND
# no one has won yet (there are empty spots)
# 1 if "X" won, 2 if "O" won, 0 if it's a draw

def is_solved(board: list):
    board = list(board)
    for row in board:
        if row == [1, 1, 1]:
            return 1
        elif row == [2, 2, 2]:
            return 2
    for column in range(len(board[0])):
        sum_column = []
        for row in range(len(board)):
            sum_column.append(board[row][column])
        if sum_column == [1, 1, 1]:
            return 1
        elif sum_column == [2, 2, 2]:
            return 2
    sum_diagonal1 = []
    for i in range(len(board)):
        sum_diagonal1.append(board[i][i])
    if sum_diagonal1 == [1, 1, 1]:
        return 1
    elif sum_diagonal1 == [2, 2, 2]:
        return 2
    sum_diagonal2 = []
    for i in range(len(board)):
        sum_diagonal2.append(board[i][len(board)-1-i])
    if sum_diagonal2 == [1, 1, 1]:
        return 1
    elif sum_diagonal2 == [2, 2, 2]:
        return 2
    for row in board:
        if 0 in row:
            return -1
    return 0
is_solved([[2, 1, 1],[0, 1, 1],[2, 2, 2]])
```

```python
# how many cakes a baker could bake considering his recipes

def cakes(recipe, available):
    min_cakes = float('inf')
    for ingredient, req_amount in recipe.items():
        if ingredient not in available:
            return 0
        possible_cakes = available[ingredient] // req_amount
        min_cakes = min(min_cakes, possible_cakes)
    return min_cakes
cakes({"flour": 500, "sugar": 200, "eggs": 1}, {"flour": 1200, "sugar": 1200, "eggs": 5, "milk": 200})
```

```python
# sum of two matrices

def matrix_addition(a, b):
    c = []
    for n in range(len(a)):
        row = []
        for m in range(len(a[0])):
            row.append(a[n][m] + b[n][m])
        c.append(row)
    return c
matrix_addition([[1, 2], [1, 2] ],[[2, 3], [2, 3]])
```

```python
# multiplization of two matrices

def matrix_mult(a: list[list[int]], b: list[list[int]]) -> list[list[int]]:
    result = []
    for i in range(len(a)):
      row = [0] * len(b[0])
      result.append(row)
    for i in range(len(a)):
        for j in range(len(b[0])):
            for k in range(len(b)):
                result[i][j] += a[i][k] * b[k][j]
    return result
matrix_mult([[1, 2], [1, 2] ],[[2, 3], [2, 3]])
```

```python
# Given a name, turn that name into a perfect square matrix

import math

def matrixfy(st):
    if len(st) == 0:
        return 'name must be at least one letter'

    initial_len = len(st)
    sqrt_initial_len = math.sqrt(initial_len)

    if sqrt_initial_len == int(sqrt_initial_len):
        n = int(sqrt_initial_len)
    else:
        n = int(math.ceil(sqrt_initial_len))
        add = (n * n) - initial_len
        st = st + '.' * add

    st_list = []
    for i in range(0, len(st), n):
        st_list.append(list(st[i:i+n]))
    return st_list
matrixfy("Beyonce")
```

```python
# a function that rotates a matrix either clockwise
# or anti-clockwise by 90 degrees, and returns the rotated array.

def rotate(matrix, direction):

    if direction == "clockwise":
        new_matrix = []

        for col in range(len(matrix[0])):
            row = []

            for r in range(len(matrix)-1, -1, -1):
                row.append(matrix[r][col])

            new_matrix.append(row)

        return new_matrix


    if direction == "counter-clockwise":
        new_matrix = []

        for col in range(len(matrix[0])-1, -1, -1):
            row = []

            for r in range(len(matrix)):
                row.append(matrix[r][col])

            new_matrix.append(row)

        return new_matrix

rotate([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
], 'clockwise')
```

```python
# a function that outputs the transpose of a matrix -
# a new matrix where the columns and rows of the original are swapped.

def transpose(matrix):
    new_matrix = []
    for col in range(len(matrix[0])):
        row = []
        for r in range(len(matrix)):
            row.append(matrix[r][col])
        new_matrix.append(row)
    return new_matrix

transpose([[1, 2, 3],[4, 5, 6]])
```

```python
# return the number of complete levels of pyramid you can make

def beeramid(bonus, price):
    if bonus == 0 or price == 0:
        return 0
    if bonus <= 0:
        return 0
    can = bonus / price
    level = 0
    while can >= level ** 2:
        can -= level ** 2
        level += 1
    return level - 1
beeramid(9, 2)
```

```python
# a function, which takes a non-negative integer (seconds) as input
# and returns the time in HH:MM:SS format

def make_readable(seconds):
    hours = seconds // 3600
    seconds = seconds % 3600
    minutes = seconds // 60
    seconds = seconds % 60
    return('{:02d}:{:02d}:{:02d}'.format(hours, minutes, seconds))
make_readable(59)
```

```python
# find the unique string

def find_uniq(arr):
    for word in arr:
        letters = set(word.lower())
        count = 0
        for other in arr:
            if set(other.lower()) == letters:
                count += 1
        if count == 1:
            return word
find_uniq([ 'Aa', 'aaa', 'aaaaa', 'BbBb', 'Aaaa', 'AaAaAa', 'a' ])
```
